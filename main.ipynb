### Yolov11

```python
from ultralytics import YOLO

model = YOLO("yolo11s-seg.pt")

results_YOLOV11 = model.train(
    data="data.yaml",
    device=0,
    epochs=150,
    patience=30,
    batch=64,
    imgsz=640,
    workers=16,
    cache=True,
    optimizer='AdamW',
    lr0=0.001,
    cos_lr=True,
    hsv_h=0.015,
    flipud=0.5,
    fliplr=0.5,
    mosaic=1.0,
    mixup=0.1,
    box=7.5,
    cls=0.5,
    rect=True,
    close_mosaic=10
)
```
